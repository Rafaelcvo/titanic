```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from collections import Counter
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier



%matplotlib inline
# %pylab inline
```

```python
train = pd.read_csv('dataset/train.csv')
test = pd.read_csv('dataset/test.csv')
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(train.corr(), annot=True, cmap="Blues")
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x='Sex', y='Survived', data= train)
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x='Pclass', y='Survived', data=train)
```

```python
train.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
test.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
train.head(2)
```

```python
train.isnull().sum()
```

```python
def transf(valor):
        if valor == 'female':
            return 1
        else:
            return 0

train['Sex'] = train['Sex'].map(transf)
test['Sex'] = test['Sex'].map(transf)
```

```python
media = np.int(train['Age'].median())
train['Age'] = train['Age'].fillna(media)
test['Age'] = test['Age'].fillna(media)
```

```python
train['Embarked'] = train['Embarked'].replace({'S':0, 'C':1, 'Q':2})
train['Embarked'] = train['Embarked'].fillna(-1)

test['Embarked'] = test['Embarked'].replace({'S':0, 'C':1, 'Q':2})
test['Embarked'] = test['Embarked'].fillna(-1)
test.isnull().sum()
```

```python
vari = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch','Fare', 'Embarked']
x = train[vari]
y = train['Survived']
x = x.fillna(-1)
y = y.fillna(-1)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
gbk = GradientBoostingClassifier()
gbk.fit(x_train, y_train)

y_pred = gbk.predict(x_val)
acc_gbk = round(accuracy_score(y_pred, y_val) * 100, 2)

print(acc_gbk)
```

```python
p_class = gbk.predict(test[vari].fillna(-1))
```

```python
tree_sub = pd.Series(p_class, index= test['PassengerId'], name='Survived')
tree_sub.to_csv("modelo_gradient1.csv", header=True)
```
