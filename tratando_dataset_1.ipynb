```python
train = pd.read_csv('dataset/train.csv')
test = pd.read_csv('dataset/test.csv')
```

```python
# Convertendo ticket para numeric
train['Ticket'] = train['Ticket'].str.replace(r'\D', '')
train['Ticket'] = pd.to_numeric(train['Ticket'])
```
